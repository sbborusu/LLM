```python
import os, sys
week1_dir = os.path.abspath(os.path.join(os.getcwd(), "..", ".."))
sys.path.insert(0, week1_dir)
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
system_prompt = """You are a famous cricket analyst. You have a deep understanding of the game and the players. 
You are also a great communicator and can explain complex cricket concepts in a simple and engaging way.

- Your task is to analyze a cricketer's biodata and summarize it in a way that is easy to understand. 
- You will be given a cricketer's biodata in the form of a file that is scraped from a website. 
- If the biodata mentions the cricketer is a future star, you should highlight it in the summary.
- If the cricket is an all-rounder, you should provide the summary of their batting and bowling.
- I would like to know the cricketer's country, batting style, bowling style, and their recent performances.
- You will provide the player's achievements and milestones in the summary.
"""

user_prompt = """Analyze the following website and provide me summary in a way that helps me understand the cricketer's abilities and achievements."""
```

```python
def generate_summary(url):
    openai = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
    # Fetch the website contents
    contents = fetch_website_contents(url)
    # Generate the summary
    response = openai.chat.completions.create(
        model="deepseek-r1:1.5b",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt + contents}
        ]
    )
    return response.choices[0].message.content
```

```python
generated_response = generate_summary('https://www.cricbuzz.com/profiles/25/sachin-tendulkar')
display(Markdown(generated_response))
```
