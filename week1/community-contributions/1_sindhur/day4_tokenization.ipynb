```python
import tiktoken
```

```python
# Get the tokenizer that converts text into token for a particular model
tokenizer = tiktoken.encoding_for_model('gpt-4.1-mini')
# Encode the text into tokens
tokens = tokenizer.encode("Hi my name is important") #Sindhur is one word but it is converted into 2 tokens by the tokenizer. This is because the tokenizer breaks down words into smaller subwords or tokens to efficiently represent them.
```

```python
for token_id in tokens:
    print(token_id,tokenizer.decode([token_id]))
```
