#Generic LLM example

```python
import os, sys
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display
```

```python
# read API key from .env file
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-4o",
    messages=[
        {
            "role": "user", 
            "content": "If a train is moving at 60 miles per hour and travels for 3 hours, how far does it go?"
        }
    ]
)
display(Markdown(response.choices[0].message.content))
# As you can see, by default, the model is able to provide some reasoning and come up with a formula, try to put it in the formula.

```

#Generic LLM example(Enable some level of thinking in the LLM via prompts)

```python
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-4o",
    messages=[
        {
            "role": "user", 
            "content": "If a train is moving at 60 miles per hour and travels for 3 hours, how far does it go? Think step by step before you give final response."
        }
    ]
)
display(Markdown(response.choices[0].message.content))
# By adding "Think step by step before you give final response", 
# we are encouraging the model to break down the problem into smaller steps, 
# which can lead to a more accurate and detailed answer.
```

#Reasoning model example (without showing the reasoning/thinking token explicitly to the user)

```python
openai = OpenAI()
response = openai.responses.create(
    model="gpt-5.4",
    reasoning={"effort": "low"},
    input=[
        {
            "role": "user", 
            "content": "If a train is moving at 60 miles per hour and travels for 3 hours, how far does it go?"
        }
    ]
)
display(Markdown(response.output_text))
#Reasoning models introduce reasoning tokens in addition to input and output tokens. 
#The models use these reasoning tokens to “think,” breaking down the prompt and considering multiple approaches to generating a response. 
#After generating reasoning tokens, the model
#The modles reasoning tokens are not visible to users, but they help the model arrive at a more accurate and detailed response.
```

#Reasoning model example (showing the reasoning/thinking token explicitly to the user)

```python
from openai import OpenAI
load_dotenv(override=True)
api_key = os.getenv('DEEPSEEK_API_KEY')
client = OpenAI(api_key=api_key, base_url="https://api.deepseek.com")
response = client.chat.completions.create(
    model="deepseek-reasoner",
    messages=[
        {
            "role": "user", 
            "content": "If a train is moving at 60 miles per hour and travels for 3 hours, how far does it go? Return your full reasoning trace including <think> tags."
        }
    ],
    stream=False
)
print(response.choices[0].message.content)
```

```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("DEEPSEEK_API_KEY")

url = "https://api.deepseek.com/v1/chat/completions"
headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}

data = {
    "model": "deepseek-reasoner",
    "messages": [
        {"role": "user", "content": "If a train is moving at 60 miles per hour and travels for 3 hours, how far does it go? Return your full reasoning trace including <think> tags."}
    ],
    "thinking": {
        "type": "enabled"
    }
}

response = requests.post(url, headers=headers, json=data)
print(response.json()["choices"][0]["message"]["content"])
```
